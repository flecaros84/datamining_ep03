### Fase 2: Comprensión de los Datos (Data Understanding)

En esta fase se realiza una exploración inicial del dataset con foco exclusivo en clientes del segmento **Residencial**. El objetivo es comprender la estructura de la base, revisar la variable objetivo, analizar la calidad de los datos y detectar patrones preliminares antes de tomar decisiones de preparación o modelamiento.

El análisis considera:

1. Carga del dataset y filtro del segmento residencial.
2. Revisión de volumen, tipos de datos y distribución de segmentos.
3. Validación de la variable objetivo `ABANDONO_FULL`.
4. Identificación de valores nulos, duplicados y variables constantes.
5. Revisión de distribuciones de variables numéricas críticas.
6. Detección preliminar de outliers mediante regla IQR y boxplots.
7. Análisis de correlación con la variable objetivo.
8. Revisión de correlaciones altas entre predictores numéricos.
9. Asociación inicial de variables categóricas con la tasa de abandono.
10. Identificación de posibles variables con fuga de información.

Este análisis servirá como base para definir qué variables conservar, transformar o excluir en la fase de preparación de datos.

```python

# ============================================================
# FASE 2 - COMPRENSIÓN DE LOS DATOS
# EDA inicial: estructura, target, nulos, outliers, correlaciones
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 120)
pd.set_option("display.float_format", "{:,.4f}".format)

# ------------------------------------------------------------
# 1. Carga del dataset y filtro residencial
# ------------------------------------------------------------

ruta_data = "data/Abandono_Clientes.csv"

df = pd.read_csv(
    ruta_data,
    sep=";",
    encoding="utf-8-sig",
    low_memory=False
)

df_res = df[df["SEGMENTO"] == "RESIDENCIAL"].copy()

target = "ABANDONO_FULL"

# ------------------------------------------------------------
# 2. Resumen general
# ------------------------------------------------------------

resumen_general = pd.DataFrame({
    "métrica": [
        "Registros dataset original",
        "Columnas dataset original",
        "Registros segmento residencial",
        "Columnas segmento residencial",
        "Participación residencial"
    ],
    "valor": [
        df.shape[0],
        df.shape[1],
        df_res.shape[0],
        df_res.shape[1],
        round(df_res.shape[0] / df.shape[0] * 100, 2)
    ]
})

print("RESUMEN GENERAL")
display(resumen_general)

distribucion_segmento = (
    df["SEGMENTO"]
    .value_counts(dropna=False)
    .rename_axis("segmento")
    .reset_index(name="registros")
)

distribucion_segmento["porcentaje"] = (
    distribucion_segmento["registros"] / len(df) * 100
).round(2)

print("DISTRIBUCIÓN POR SEGMENTO")
display(distribucion_segmento)

tipos_datos = (
    df_res.dtypes.astype(str)
    .value_counts()
    .rename_axis("tipo_dato")
    .reset_index(name="cantidad_variables")
)

print("TIPOS DE DATOS EN SEGMENTO RESIDENCIAL")
display(tipos_datos)

# ------------------------------------------------------------
# 3. Distribución de la variable objetivo
# ------------------------------------------------------------

distribucion_target = (
    df_res[target]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis(target)
    .reset_index(name="registros")
)

distribucion_target["porcentaje"] = (
    distribucion_target["registros"] / len(df_res) * 100
).round(2)

print("DISTRIBUCIÓN DE VARIABLE OBJETIVO")
display(distribucion_target)

plt.figure(figsize=(6, 4))
plt.bar(distribucion_target[target].astype(str), distribucion_target["registros"])
plt.title("Distribución de ABANDONO_FULL")
plt.xlabel("Clase")
plt.ylabel("Registros")
plt.show()

# ------------------------------------------------------------
# 4. Duplicados, nulos y cardinalidad
# ------------------------------------------------------------

duplicados_completos = df_res.duplicated().sum()

nulos = df_res.isna().sum().to_frame("nulos")
nulos["porcentaje_nulos"] = (nulos["nulos"] / len(df_res) * 100).round(2)
nulos = nulos[nulos["nulos"] > 0].sort_values("porcentaje_nulos", ascending=False)

cardinalidad = pd.DataFrame({
    "variable": df_res.columns,
    "tipo_dato": df_res.dtypes.astype(str).values,
    "valores_unicos": [df_res[col].nunique(dropna=False) for col in df_res.columns],
    "porcentaje_unicos": [
        round(df_res[col].nunique(dropna=False) / len(df_res) * 100, 4)
        for col in df_res.columns
    ]
}).sort_values("valores_unicos")

variables_constantes = cardinalidad[cardinalidad["valores_unicos"] == 1]["variable"].tolist()
variables_100_nulas = nulos[nulos["porcentaje_nulos"] == 100].index.tolist()

resumen_calidad = pd.DataFrame({
    "métrica": [
        "Duplicados completos",
        "Variables con nulos",
        "Variables constantes",
        "Variables 100% nulas",
        "Valores únicos en RUT_CLIENTE"
    ],
    "valor": [
        duplicados_completos,
        len(nulos),
        len(variables_constantes),
        len(variables_100_nulas),
        df_res["RUT_CLIENTE"].nunique(dropna=False) if "RUT_CLIENTE" in df_res.columns else np.nan
    ]
})

print("RESUMEN DE CALIDAD")
display(resumen_calidad)

print("VARIABLES CON MAYOR PORCENTAJE DE NULOS")
display(nulos.head(25))

print("VARIABLES CONSTANTES")
display(pd.DataFrame({"variable": variables_constantes}))

print("VARIABLES 100% NULAS")
display(pd.DataFrame({"variable": variables_100_nulas}))

# ------------------------------------------------------------
# 5. Revisión de variables críticas numéricas
# ------------------------------------------------------------

variables_numericas_criticas = [
    "SS_INTE",
    "SS_TV",
    "SS_TEL",
    "SS_SVA",
    "SS_MOVIL",
    "SS_DATOS",
    "SS_TV_APP",
    "ANTG_MESES_TODO",
    "MONTO_POR_SS",
    "VLOR_NETO_PROM",
    "MESES_COBROS",
    "VELOCIDAD_BAJADA",
    "VELOCIDAD_SUBIDA",
    "NUM_RECLAMOS",
    "NUM_RECLAMOS_TECNICO_TERRENO",
    "EVENTOS_DISTINTOS",
    "HORAS_TOTALES_RUT",
    "SUPERFICIE_KM2",
    "LATITUD_DECIMAL",
    "LONGITUD_DECIMAL"
]

variables_numericas_criticas = [
    col for col in variables_numericas_criticas
    if col in df_res.columns
]

resumen_numericas = df_res[variables_numericas_criticas].describe().T

resumen_numericas["nulos"] = df_res[variables_numericas_criticas].isna().sum()
resumen_numericas["porcentaje_nulos"] = (
    df_res[variables_numericas_criticas].isna().mean() * 100
).round(2)
resumen_numericas["asimetria"] = df_res[variables_numericas_criticas].skew(numeric_only=True)
resumen_numericas["curtosis"] = df_res[variables_numericas_criticas].kurtosis(numeric_only=True)

print("RESUMEN ESTADÍSTICO DE VARIABLES NUMÉRICAS CRÍTICAS")
display(resumen_numericas)

for col in variables_numericas_criticas:
    plt.figure(figsize=(7, 4))
    plt.hist(df_res[col].dropna(), bins=40)
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()

# ------------------------------------------------------------
# 6. Outliers por regla IQR
# ------------------------------------------------------------

outliers_resumen = []

for col in variables_numericas_criticas:
    serie = df_res[col].dropna()
    
    if len(serie) == 0:
        continue
    
    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)
    iqr = q3 - q1
    
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    outliers = ((serie < limite_inferior) | (serie > limite_superior)).sum()
    
    outliers_resumen.append({
        "variable": col,
        "q1": q1,
        "q3": q3,
        "iqr": iqr,
        "limite_inferior": limite_inferior,
        "limite_superior": limite_superior,
        "outliers_iqr": outliers,
        "porcentaje_outliers": round(outliers / len(serie) * 100, 2)
    })

outliers_resumen_df = pd.DataFrame(outliers_resumen).sort_values(
    "porcentaje_outliers",
    ascending=False
)

print("RESUMEN DE OUTLIERS POR REGLA IQR")
display(outliers_resumen_df)

for col in variables_numericas_criticas:
    plt.figure(figsize=(7, 3))
    plt.boxplot(df_res[col].dropna(), vert=False)
    plt.title(f"Boxplot de {col}")
    plt.xlabel(col)
    plt.show()

# ------------------------------------------------------------
# 7. Validación de abandono por servicio
# ------------------------------------------------------------

pares_servicio_abandono = {
    "Internet": ("SS_INTE", "ABANDONO_FULL_INTE", "ABANDONO_INTE"),
    "TV": ("SS_TV", "ABANDONO_FULL_TV", "ABANDONO_TV"),
    "Telefonía": ("SS_TEL", "ABANDONO_FULL_TEL", "ABANDONO_TEL"),
    "TV App": ("SS_TV_APP", "ABANDONO_FULL_TV_APP", "ABANDONO_TV_APP"),
    "SVA": ("SS_SVA", "ABANDONO_FULL_SVA", "ABANDONO_SVA")
}

validacion_abandono_servicio = []

for servicio, columnas in pares_servicio_abandono.items():
    col_servicio, col_full, col_parcial = columnas
    
    if col_servicio in df_res.columns and col_full in df_res.columns:
        sin_servicio = df_res[df_res[col_servicio] == 0]
        
        if len(sin_servicio) > 0:
            sin_servicio_marcados_full = sin_servicio[col_full].sum()
            porcentaje_sin_servicio_marcado_full = (
                sin_servicio_marcados_full / len(sin_servicio) * 100
            )
        else:
            sin_servicio_marcados_full = 0
            porcentaje_sin_servicio_marcado_full = 0
        
        validacion_abandono_servicio.append({
            "servicio": servicio,
            "variable_servicio": col_servicio,
            "variable_abandono_full": col_full,
            "clientes_sin_servicio": len(sin_servicio),
            "sin_servicio_marcados_full": int(sin_servicio_marcados_full),
            "porcentaje_sin_servicio_marcado_full": round(porcentaje_sin_servicio_marcado_full, 2),
            "interpretacion": "Mezcla no contratación con abandono" if porcentaje_sin_servicio_marcado_full > 50 else "Consistente para revisión exploratoria"
        })

validacion_abandono_servicio_df = pd.DataFrame(validacion_abandono_servicio)

print("VALIDACIÓN DE VARIABLES ABANDONO_FULL POR SERVICIO")
display(validacion_abandono_servicio_df)

# ------------------------------------------------------------
# 8. Correlación con target
# ------------------------------------------------------------

df_corr = df_res.copy()

# Para correlación, se imputan temporalmente numéricas con mediana.
# Esto es solo para análisis exploratorio, no para modelamiento final.
columnas_numericas = df_corr.select_dtypes(include=["int64", "float64"]).columns.tolist()

for col in columnas_numericas:
    if df_corr[col].isna().any():
        df_corr[col] = df_corr[col].fillna(df_corr[col].median())

correlacion_target = (
    df_corr[columnas_numericas]
    .corr(method="pearson")[target]
    .drop(target)
    .sort_values(key=lambda x: x.abs(), ascending=False)
    .reset_index()
)

correlacion_target.columns = ["variable", "correlacion_con_abandono"]
correlacion_target["correlacion_abs"] = correlacion_target["correlacion_con_abandono"].abs()

print("TOP 20 VARIABLES NUMÉRICAS MÁS CORRELACIONADAS CON ABANDONO_FULL")
display(correlacion_target.head(20))

top_corr = correlacion_target.head(15).sort_values("correlacion_con_abandono")

plt.figure(figsize=(8, 6))
plt.barh(top_corr["variable"], top_corr["correlacion_con_abandono"])
plt.title("Top correlaciones numéricas con ABANDONO_FULL")
plt.xlabel("Correlación")
plt.ylabel("Variable")
plt.show()

# ------------------------------------------------------------
# 9. Correlaciones altas entre predictores numéricos
# ------------------------------------------------------------

columnas_numericas_predictoras = [
    col for col in columnas_numericas
    if col != target and not col.startswith("ABANDONO_") and not col.startswith("ABA_")
]

matriz_corr = df_corr[columnas_numericas_predictoras].corr(method="pearson").abs()

pares_corr_altos = (
    matriz_corr
    .where(np.triu(np.ones(matriz_corr.shape), k=1).astype(bool))
    .stack()
    .reset_index()
)

pares_corr_altos.columns = ["variable_1", "variable_2", "correlacion_abs"]

pares_corr_altos = pares_corr_altos[
    pares_corr_altos["correlacion_abs"] >= 0.80
].sort_values("correlacion_abs", ascending=False)

print("PARES DE VARIABLES NUMÉRICAS CON CORRELACIÓN ALTA >= 0.80")
display(pares_corr_altos)

# Heatmap acotado para variables numéricas más asociadas al target
variables_heatmap = correlacion_target.head(12)["variable"].tolist()
variables_heatmap = [
    col for col in variables_heatmap
    if col in columnas_numericas_predictoras
]

if len(variables_heatmap) >= 2:
    matriz_heatmap = df_corr[variables_heatmap + [target]].corr()

    plt.figure(figsize=(10, 8))
    plt.imshow(matriz_heatmap, aspect="auto")
    plt.colorbar(label="Correlación")
    plt.xticks(
        ticks=range(len(matriz_heatmap.columns)),
        labels=matriz_heatmap.columns,
        rotation=90
    )
    plt.yticks(
        ticks=range(len(matriz_heatmap.index)),
        labels=matriz_heatmap.index
    )
    plt.title("Matriz de correlación acotada")
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 10. Asociación categórica con abandono
# ------------------------------------------------------------

columnas_categoricas_clave = [
    "EMPRESA",
    "CRUCE_ACTIVO",
    "CODI_ADMZONAL",
    "CODI_LOCALIDA",
    "CONDICION_ESPECIAL",
    "PRIORITARIO",
    "COMUNA",
    "PROVINCIA",
    "REGION"
]

columnas_categoricas_clave = [
    col for col in columnas_categoricas_clave
    if col in df_res.columns
]

resumen_categoricas = []

for col in columnas_categoricas_clave:
    tabla = (
        df_res
        .assign(**{col: df_res[col].fillna("SIN_INFORMAR")})
        .groupby(col)[target]
        .agg(["count", "mean"])
        .reset_index()
        .rename(columns={
            "count": "registros",
            "mean": "tasa_abandono"
        })
    )
    
    tabla["tasa_abandono"] = (tabla["tasa_abandono"] * 100).round(2)
    
    resumen_categoricas.append({
        "variable": col,
        "categorias": tabla.shape[0],
        "registros_categoria_min": int(tabla["registros"].min()),
        "registros_categoria_max": int(tabla["registros"].max()),
        "tasa_abandono_min": tabla["tasa_abandono"].min(),
        "tasa_abandono_max": tabla["tasa_abandono"].max(),
        "rango_tasa_abandono": round(
            tabla["tasa_abandono"].max() - tabla["tasa_abandono"].min(),
            2
        )
    })

resumen_categoricas_df = pd.DataFrame(resumen_categoricas).sort_values(
    "rango_tasa_abandono",
    ascending=False
)

print("RESUMEN DE ASOCIACIÓN CATEGÓRICA CON ABANDONO")
display(resumen_categoricas_df)

for col in columnas_categoricas_clave:
    print(f"TASA DE ABANDONO POR {col}")
    
    tabla_categoria = (
        df_res
        .assign(**{col: df_res[col].fillna("SIN_INFORMAR")})
        .groupby(col)[target]
        .agg(["count", "mean"])
        .reset_index()
        .rename(columns={
            "count": "registros",
            "mean": "tasa_abandono"
        })
    )
    
    tabla_categoria["tasa_abandono"] = (
        tabla_categoria["tasa_abandono"] * 100
    ).round(2)
    
    tabla_categoria = tabla_categoria.sort_values(
        ["tasa_abandono", "registros"],
        ascending=[False, False]
    )
    
    display(tabla_categoria.head(15))

# ------------------------------------------------------------
# 11. Variables candidatas a fuga de información
# ------------------------------------------------------------

variables_aba = [
    col for col in df_res.columns
    if col.startswith("ABA_")
]

variables_abandono_no_target = [
    col for col in df_res.columns
    if col.startswith("ABANDONO_") and col != target
]

variables_gestion_comercial = [
    col for col in [
        "ESTADO_CONTACTO",
        "FIDELIZADO",
        "DERIVADO_A_EJECUTIVO",
        "FECHA_CONTACTO"
    ]
    if col in df_res.columns
]

resumen_leakage = pd.DataFrame([
    {
        "grupo": "Variables ABA_*",
        "cantidad": len(variables_aba),
        "variables": variables_aba,
        "criterio": "Excluir por contener información posterior o directamente relacionada al abandono."
    },
    {
        "grupo": "Variables ABANDONO_* distintas del target",
        "cantidad": len(variables_abandono_no_target),
        "variables": variables_abandono_no_target,
        "criterio": "Excluir como predictores para evitar fuga de información."
    },
    {
        "grupo": "Gestión comercial",
        "cantidad": len(variables_gestion_comercial),
        "variables": variables_gestion_comercial,
        "criterio": "Revisar temporalidad; excluir si corresponde a acciones posteriores al riesgo o abandono."
    }
])

print("RESUMEN DE POSIBLE FUGA DE INFORMACIÓN")
display(resumen_leakage)
```

#### Matriz de correlación de variables numéricas críticas

Se calcula una matriz de correlación entre las variables numéricas críticas revisadas en el análisis exploratorio. Esta matriz permite observar relaciones lineales entre variables de servicios contratados, antigüedad, facturación, reclamos, eventos técnicos, ubicación geográfica y la variable objetivo `ABANDONO_FULL`.

El objetivo de esta revisión es identificar patrones de asociación relevantes y detectar posibles redundancias entre predictores antes de la preparación final y el modelamiento.

```python

# ============================================================
# MATRIZ DE CORRELACIÓN ENTRE VARIABLES NUMÉRICAS CRÍTICAS
# ============================================================

import numpy as np
import matplotlib.pyplot as plt

# Variables numéricas críticas revisadas en el EDA
variables_matriz_corr = [
    "ABANDONO_FULL",
    "SS_INTE",
    "SS_TV",
    "SS_TEL",
    "SS_SVA",
    "SS_MOVIL",
    "SS_DATOS",
    "SS_TV_APP",
    "ANTG_MESES_TODO",
    "MONTO_POR_SS",
    "VLOR_NETO_PROM",
    "MESES_COBROS",
    "VELOCIDAD_BAJADA",
    "VELOCIDAD_SUBIDA",
    "NUM_RECLAMOS",
    "NUM_RECLAMOS_TECNICO_TERRENO",
    "EVENTOS_DISTINTOS",
    "HORAS_TOTALES_RUT",
    "SUPERFICIE_KM2",
    "LATITUD_DECIMAL",
    "LONGITUD_DECIMAL"
]

variables_matriz_corr = [
    col for col in variables_matriz_corr
    if col in df_res.columns
]

# Copia auxiliar solo para análisis exploratorio de correlación
df_corr_criticas = df_res[variables_matriz_corr].copy()

# Imputación temporal solo para calcular correlaciones exploratorias.
# No reemplaza la preparación formal del dataset.
for col in df_corr_criticas.columns:
    if df_corr_criticas[col].isna().any():
        df_corr_criticas[col] = df_corr_criticas[col].fillna(df_corr_criticas[col].median())

# Matriz de correlación de Pearson
matriz_correlacion_criticas = df_corr_criticas.corr(method="pearson")

print("MATRIZ DE CORRELACIÓN - VARIABLES NUMÉRICAS CRÍTICAS")
display(matriz_correlacion_criticas)

# Heatmap completo
plt.figure(figsize=(14, 10))
plt.imshow(
    matriz_correlacion_criticas,
    aspect="auto",
    vmin=-1,
    vmax=1
)
plt.colorbar(label="Correlación de Pearson")

plt.xticks(
    ticks=range(len(matriz_correlacion_criticas.columns)),
    labels=matriz_correlacion_criticas.columns,
    rotation=90
)

plt.yticks(
    ticks=range(len(matriz_correlacion_criticas.index)),
    labels=matriz_correlacion_criticas.index
)

plt.title("Matriz de correlación de variables numéricas críticas")
plt.tight_layout()
plt.show()

# Pares de correlación ordenados, excluyendo diagonal y duplicados
pares_correlacion_criticas = (
    matriz_correlacion_criticas
    .where(np.triu(np.ones(matriz_correlacion_criticas.shape), k=1).astype(bool))
    .stack()
    .reset_index()
)

pares_correlacion_criticas.columns = [
    "variable_1",
    "variable_2",
    "correlacion"
]

pares_correlacion_criticas["correlacion_abs"] = (
    pares_correlacion_criticas["correlacion"].abs()
)

pares_correlacion_criticas = pares_correlacion_criticas.sort_values(
    "correlacion_abs",
    ascending=False
)

print("TOP 20 PARES DE VARIABLES NUMÉRICAS CRÍTICAS MÁS CORRELACIONADAS")
display(pares_correlacion_criticas.head(20))
```

### Fase 3: Preparación de los Datos (Data Preparation)

A partir de los hallazgos de la fase exploratoria, se prepara el dataset para análisis avanzado y modelamiento predictivo. Esta etapa busca construir una base limpia, consistente y sin fuga de información, manteniendo variables comerciales, técnicas, geográficas y de servicios que puedan aportar señal para anticipar abandono residencial.

Las principales decisiones de preparación son:

1. Eliminar duplicados completos.
2. Excluir variables constantes, completamente nulas o no informativas.
3. Excluir variables `ABA_*` y variables `ABANDONO_*` distintas de `ABANDONO_FULL`, para evitar fuga de información.
4. Excluir variables de gestión comercial con temporalidad ambigua, como `FIDELIZADO`, `ESTADO_CONTACTO`, `DERIVADO_A_EJECUTIVO` y `FECHA_CONTACTO`.
5. Reemplazar campos tipo lista por variables binarias `FLAG_*`, conservando la señal de existencia de registros sin introducir alta cardinalidad.
6. Imputar valores nulos según el tipo de variable.
7. Mantener outliers como observaciones válidas, dado que en su mayoría corresponden a fenómenos reales del negocio, como diferencias de facturación, antigüedad, reclamos, velocidad o ubicación.
8. Generar el dataset preparado para modelamiento, separando predictores y variable objetivo.

```python

# ============================================================
# FASE 3 - PREPARACIÓN DE LOS DATOS
# Preparación final basada en el EDA
# ============================================================

import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 120)
pd.set_option("display.float_format", "{:,.4f}".format)

target = "ABANDONO_FULL"

# ------------------------------------------------------------
# 1. Carga segura del dataset residencial
# ------------------------------------------------------------

try:
    df_res
except NameError:
    ruta_data = "data/Abandono_Clientes.csv"
    df = pd.read_csv(
        ruta_data,
        sep=";",
        encoding="utf-8-sig",
        low_memory=False
    )
    df_res = df[df["SEGMENTO"] == "RESIDENCIAL"].copy()

registros_iniciales = len(df_res)
columnas_iniciales = df_res.shape[1]

# ------------------------------------------------------------
# 2. Eliminación de duplicados completos
# ------------------------------------------------------------

df_prep = df_res.drop_duplicates().copy()

duplicados_eliminados = registros_iniciales - len(df_prep)

# ------------------------------------------------------------
# 3. Variables derivadas simples
# ------------------------------------------------------------

# Campos tipo lista: se transforman en flags para evitar alta cardinalidad.
columnas_lista = [
    "LISTA_SS_SVA",
    "LISTA_CAUSAS_COMERCIAL_NO_TERRENO",
    "LISTA_CAUSAS_TECNICO_NO_TERRENO",
    "LISTA_CAUSAS_TECNICO_TERRENO",
    "LISTA_TIPO_RECLAMOS_TECNICO_NO_TERRENO",
    "LISTA_TIPO_RECLAMOS_TECNICO_TERRENO"
]

for col in columnas_lista:
    if col in df_prep.columns:
        df_prep[f"FLAG_{col}"] = df_prep[col].notna().astype(int)

# Conteos de reclamos: nulo se interpreta como ausencia de registro observado.
for col in ["NUM_RECLAMOS", "NUM_RECLAMOS_TECNICO_TERRENO"]:
    if col in df_prep.columns:
        df_prep[col] = df_prep[col].fillna(0)

if "NUM_RECLAMOS" in df_prep.columns:
    df_prep["FLAG_TIENE_RECLAMOS"] = (df_prep["NUM_RECLAMOS"] > 0).astype(int)

if "EVENTOS_DISTINTOS" in df_prep.columns:
    df_prep["FLAG_TIENE_EVENTOS_TECNICOS"] = (df_prep["EVENTOS_DISTINTOS"] > 0).astype(int)

# Paquetización: suma de servicios principales.
columnas_servicios = [
    "SS_INTE",
    "SS_TV",
    "SS_TEL",
    "SS_SVA",
    "SS_MOVIL",
    "SS_DATOS",
    "SS_TV_APP"
]

columnas_servicios_existentes = [
    col for col in columnas_servicios
    if col in df_prep.columns
]

if columnas_servicios_existentes:
    df_prep["TOTAL_SERVICIOS"] = df_prep[columnas_servicios_existentes].sum(axis=1)

# ------------------------------------------------------------
# 4. Definición de columnas a excluir
# ------------------------------------------------------------

columnas_constantes = [
    col for col in df_prep.columns
    if df_prep[col].nunique(dropna=False) == 1
]

columnas_100_nulas = [
    col for col in df_prep.columns
    if df_prep[col].isna().mean() == 1
]

columnas_aba = [
    col for col in df_prep.columns
    if col.startswith("ABA_")
]

columnas_abandono_no_target = [
    col for col in df_prep.columns
    if col.startswith("ABANDONO_") and col != target
]

columnas_identificacion = [
    "RUT_CLIENTE",
    "NOMB_RAZON",
    "SEGMENTO",
    "FECHA"
]

columnas_fechas_crudas = [
    "FECH_ALTA",
    "FECHA_MIN",
    "MES_RECLAMO_MIN",
    "MES_RECLAMO_MAX"
]

columnas_gestion_comercial = [
    "ESTADO_CONTACTO",
    "FIDELIZADO",
    "DERIVADO_A_EJECUTIVO",
    "FECHA_CONTACTO"
]

columnas_listas_originales = columnas_lista

columnas_excluir = sorted(set(
    columnas_constantes
    + columnas_100_nulas
    + columnas_aba
    + columnas_abandono_no_target
    + columnas_identificacion
    + columnas_fechas_crudas
    + columnas_gestion_comercial
    + columnas_listas_originales
))

columnas_excluir = [
    col for col in columnas_excluir
    if col in df_prep.columns and col != target
]

# ------------------------------------------------------------
# 5. Construcción del dataset preparado
# ------------------------------------------------------------

df_model = df_prep.drop(columns=columnas_excluir, errors="ignore").copy()

# Validación de target
df_model = df_model[df_model[target].notna()].copy()
df_model[target] = df_model[target].astype(int)

# ------------------------------------------------------------
# 6. Imputación de valores nulos
# ------------------------------------------------------------

columnas_numericas = df_model.select_dtypes(include=["int64", "float64"]).columns.tolist()
columnas_categoricas = df_model.select_dtypes(include=["object", "string"]).columns.tolist()

columnas_numericas_predictoras = [
    col for col in columnas_numericas
    if col != target
]

for col in columnas_numericas_predictoras:
    df_model[col] = df_model[col].fillna(df_model[col].median())

for col in columnas_categoricas:
    df_model[col] = df_model[col].fillna("SIN_INFORMAR")

# ------------------------------------------------------------
# 7. Separación X/y y partición train-test
# ------------------------------------------------------------

X = df_model.drop(columns=[target])
y = df_model[target]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

# ------------------------------------------------------------
# 8. Artefactos de control
# ------------------------------------------------------------

resumen_preparacion = pd.DataFrame({
    "métrica": [
        "Registros iniciales residenciales",
        "Duplicados completos eliminados",
        "Registros finales preparados",
        "Columnas iniciales",
        "Columnas excluidas",
        "Columnas finales incluyendo target",
        "Variables predictoras finales",
        "Variables numéricas predictoras",
        "Variables categóricas predictoras",
        "Nulos restantes en dataset preparado",
        "Casos positivos target",
        "Porcentaje positivo target",
        "Registros train",
        "Registros test",
        "Porcentaje positivo train",
        "Porcentaje positivo test"
    ],
    "valor": [
        registros_iniciales,
        duplicados_eliminados,
        len(df_model),
        columnas_iniciales,
        len(columnas_excluir),
        df_model.shape[1],
        X.shape[1],
        len([col for col in X.columns if col in columnas_numericas_predictoras]),
        len([col for col in X.columns if col in columnas_categoricas]),
        int(df_model.isna().sum().sum()),
        int(y.sum()),
        round(y.mean() * 100, 2),
        len(X_train),
        len(X_test),
        round(y_train.mean() * 100, 2),
        round(y_test.mean() * 100, 2)
    ]
})

criterios_exclusion = pd.DataFrame([
    {
        "criterio": "Variables constantes",
        "cantidad": len([col for col in columnas_constantes if col in columnas_excluir]),
        "columnas": [col for col in columnas_constantes if col in columnas_excluir]
    },
    {
        "criterio": "Variables 100% nulas",
        "cantidad": len([col for col in columnas_100_nulas if col in columnas_excluir]),
        "columnas": [col for col in columnas_100_nulas if col in columnas_excluir]
    },
    {
        "criterio": "Variables ABA_*",
        "cantidad": len([col for col in columnas_aba if col in columnas_excluir]),
        "columnas": [col for col in columnas_aba if col in columnas_excluir]
    },
    {
        "criterio": "Variables ABANDONO_* no target",
        "cantidad": len([col for col in columnas_abandono_no_target if col in columnas_excluir]),
        "columnas": [col for col in columnas_abandono_no_target if col in columnas_excluir]
    },
    {
        "criterio": "Identificación / no informativas",
        "cantidad": len([col for col in columnas_identificacion if col in columnas_excluir]),
        "columnas": [col for col in columnas_identificacion if col in columnas_excluir]
    },
    {
        "criterio": "Fechas crudas",
        "cantidad": len([col for col in columnas_fechas_crudas if col in columnas_excluir]),
        "columnas": [col for col in columnas_fechas_crudas if col in columnas_excluir]
    },
    {
        "criterio": "Gestión comercial con temporalidad ambigua",
        "cantidad": len([col for col in columnas_gestion_comercial if col in columnas_excluir]),
        "columnas": [col for col in columnas_gestion_comercial if col in columnas_excluir]
    },
    {
        "criterio": "Listas originales reemplazadas por flags",
        "cantidad": len([col for col in columnas_listas_originales if col in columnas_excluir]),
        "columnas": [col for col in columnas_listas_originales if col in columnas_excluir]
    }
])

distribucion_train_test = pd.DataFrame({
    "partición": ["Train", "Test"],
    "registros": [len(y_train), len(y_test)],
    "positivos": [int(y_train.sum()), int(y_test.sum())],
    "porcentaje_positivos": [
        round(y_train.mean() * 100, 2),
        round(y_test.mean() * 100, 2)
    ]
})

columnas_finales = pd.DataFrame({
    "columna": df_model.columns,
    "tipo_dato": df_model.dtypes.astype(str).values
})

print("RESUMEN DE PREPARACIÓN")
display(resumen_preparacion)

print("CRITERIOS DE EXCLUSIÓN")
display(criterios_exclusion)

print("DISTRIBUCIÓN TRAIN / TEST")
display(distribucion_train_test)

print("COLUMNAS FINALES DEL DATASET PREPARADO")
display(columnas_finales)

# ------------------------------------------------------------
# 9. Guardado del dataset preparado
# ------------------------------------------------------------

os.makedirs("data/processed", exist_ok=True)

ruta_salida = "data/processed/clientes_residenciales_preparado.csv"

df_model.to_csv(
    ruta_salida,
    index=False,
    encoding="utf-8-sig"
)

print(f"Dataset preparado guardado en: {ruta_salida}")
```

### Fase 4: Modelamiento

En esta fase se aplican modelos supervisados y no supervisados sobre el dataset preparado de clientes residenciales. El objetivo es abordar el problema desde tres enfoques complementarios:

1. **Aprendizaje supervisado:** predecir la probabilidad de abandono mediante modelos de clasificación.
2. **Aprendizaje no supervisado:** segmentar clientes según características comerciales, técnicas, geográficas y de servicios.
3. **Reglas de asociación:** descubrir patrones de contratación conjunta de servicios que puedan apoyar estrategias de retención y venta cruzada.

Para el aprendizaje supervisado se utiliza como variable objetivo `ABANDONO_FULL`. Dado que la clase positiva representa aproximadamente el **3,49%** de los registros, se considera un problema de clasificación desbalanceada. Por ello, la evaluación prioriza métricas como `Recall`, `Precision`, `F1-Score`, `ROC-AUC` y `PR-AUC`, además de analizar la matriz de confusión desde una perspectiva de negocio.

Las variables numéricas son escaladas mediante `StandardScaler` o `MinMaxScaler`, según el modelo, y las variables categóricas son transformadas mediante `OneHotEncoder`. Estas transformaciones se aplican dentro de pipelines para evitar fuga de información.

Los modelos supervisados evaluados son:

- Regresión Logística balanceada.
- Árbol de Decisión balanceado.
- Naive Bayes.
- SVM lineal balanceado como modelo adicional investigado.

Para el aprendizaje no supervisado se aplican:

- K-Means.
- Clustering Jerárquico sobre una muestra representativa, debido al tamaño de la base.

Finalmente, se aplican reglas de asociación tipo Apriori sobre variables de servicios contratados y servicios vigentes, con el objetivo de identificar combinaciones de servicios que puedan actuar como anclas de fidelización.

```python
# ============================================================
# FASE 4 - MODELAMIENTO
# Supervisado, No Supervisado y Reglas de Asociación
# ============================================================

import os
import itertools
import warnings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

from sklearn.cluster import MiniBatchKMeans, AgglomerativeClustering
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
    classification_report,
    silhouette_score
)

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 120)
pd.set_option("display.float_format", "{:,.4f}".format)

target = "ABANDONO_FULL"

# ------------------------------------------------------------
# 1. Carga segura del dataset preparado
# ------------------------------------------------------------

try:
    df_model
except NameError:
    ruta_preparado = "data/processed/clientes_residenciales_preparado.csv"
    df_model = pd.read_csv(ruta_preparado, encoding="utf-8-sig")

X = df_model.drop(columns=[target])
y = df_model[target].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

columnas_numericas = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
columnas_categoricas = X_train.select_dtypes(include=["object", "string"]).columns.tolist()

try:
    encoder_sparse = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
except TypeError:
    encoder_sparse = OneHotEncoder(handle_unknown="ignore", sparse=True)

# Preprocesamiento para modelos lineales y árbol
preprocesamiento_standard = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), columnas_numericas),
        ("cat", encoder_sparse, columnas_categoricas)
    ],
    remainder="drop"
)

# Preprocesamiento para Naive Bayes: requiere variables no negativas
preprocesamiento_minmax = ColumnTransformer(
    transformers=[
        ("num", MinMaxScaler(), columnas_numericas),
        ("cat", encoder_sparse, columnas_categoricas)
    ],
    remainder="drop"
)

resumen_modelamiento = pd.DataFrame({
    "métrica": [
        "Registros entrenamiento",
        "Registros prueba",
        "Variables predictoras originales",
        "Variables numéricas",
        "Variables categóricas",
        "Casos positivos train",
        "Casos positivos test",
        "Porcentaje abandono train",
        "Porcentaje abandono test"
    ],
    "valor": [
        len(X_train),
        len(X_test),
        X_train.shape[1],
        len(columnas_numericas),
        len(columnas_categoricas),
        int(y_train.sum()),
        int(y_test.sum()),
        round(y_train.mean() * 100, 2),
        round(y_test.mean() * 100, 2)
    ]
})

print("RESUMEN DEL SET DE MODELAMIENTO")
display(resumen_modelamiento)

# ============================================================
# 2. MODELOS SUPERVISADOS
# ============================================================

modelos_supervisados = {
    "Regresión logística balanceada": Pipeline(steps=[
        ("preprocesamiento", preprocesamiento_standard),
        ("modelo", LogisticRegression(
            class_weight="balanced",
            max_iter=1000,
            solver="liblinear",
            random_state=42
        ))
    ]),
    "Árbol de decisión balanceado": Pipeline(steps=[
        ("preprocesamiento", preprocesamiento_standard),
        ("modelo", DecisionTreeClassifier(
            class_weight="balanced",
            max_depth=6,
            min_samples_leaf=100,
            random_state=42
        ))
    ]),
    "Naive Bayes": Pipeline(steps=[
        ("preprocesamiento", preprocesamiento_minmax),
        ("modelo", MultinomialNB())
    ]),
    "SVM lineal balanceado": Pipeline(steps=[
        ("preprocesamiento", preprocesamiento_standard),
        ("modelo", LinearSVC(
            class_weight="balanced",
            max_iter=5000,
            random_state=42
        ))
    ])
}

resultados_supervisados = []
modelos_entrenados = {}

for nombre_modelo, pipeline in modelos_supervisados.items():
    pipeline.fit(X_train, y_train)
    
    y_pred = pipeline.predict(X_test)
    
    if hasattr(pipeline.named_steps["modelo"], "predict_proba"):
        y_score = pipeline.predict_proba(X_test)[:, 1]
    elif hasattr(pipeline.named_steps["modelo"], "decision_function"):
        y_score = pipeline.decision_function(X_test)
    else:
        y_score = y_pred
    
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    
    resultados_supervisados.append({
        "modelo": nombre_modelo,
        "accuracy": round(accuracy_score(y_test, y_pred), 4),
        "precision": round(precision_score(y_test, y_pred, zero_division=0), 4),
        "recall": round(recall_score(y_test, y_pred, zero_division=0), 4),
        "f1_score": round(f1_score(y_test, y_pred, zero_division=0), 4),
        "roc_auc": round(roc_auc_score(y_test, y_score), 4),
        "pr_auc": round(average_precision_score(y_test, y_score), 4),
        "tn": tn,
        "fp": fp,
        "fn": fn,
        "tp": tp
    })
    
    modelos_entrenados[nombre_modelo] = {
        "pipeline": pipeline,
        "y_pred": y_pred,
        "y_score": y_score
    }

resultados_supervisados_df = pd.DataFrame(resultados_supervisados).sort_values(
    by=["f1_score", "recall", "pr_auc"],
    ascending=False
)

print("RESULTADOS COMPARATIVOS - MODELOS SUPERVISADOS")
display(resultados_supervisados_df)

mejor_modelo_nombre = resultados_supervisados_df.iloc[0]["modelo"]
mejor_modelo = modelos_entrenados[mejor_modelo_nombre]

print(f"MEJOR MODELO SUPERVISADO SEGÚN F1-SCORE: {mejor_modelo_nombre}")
print("-" * 70)

print(classification_report(
    y_test,
    mejor_modelo["y_pred"],
    target_names=["No abandono", "Abandono"],
    zero_division=0
))

matriz_confusion_mejor = pd.DataFrame(
    confusion_matrix(y_test, mejor_modelo["y_pred"]),
    index=["Real no abandono", "Real abandono"],
    columns=["Predicho no abandono", "Predicho abandono"]
)

print("MATRIZ DE CONFUSIÓN DEL MEJOR MODELO SUPERVISADO")
display(matriz_confusion_mejor)

# Importancia de variables del árbol de decisión
nombre_arbol = "Árbol de decisión balanceado"

pipeline_arbol = modelos_entrenados[nombre_arbol]["pipeline"]
modelo_arbol = pipeline_arbol.named_steps["modelo"]
preprocessor_arbol = pipeline_arbol.named_steps["preprocesamiento"]

nombres_num = columnas_numericas
nombres_cat = (
    preprocessor_arbol
    .named_transformers_["cat"]
    .get_feature_names_out(columnas_categoricas)
    .tolist()
)

nombres_features_arbol = nombres_num + nombres_cat

importancia_arbol = pd.DataFrame({
    "variable": nombres_features_arbol,
    "importancia": modelo_arbol.feature_importances_
}).sort_values("importancia", ascending=False)

print("TOP 20 VARIABLES MÁS IMPORTANTES SEGÚN ÁRBOL DE DECISIÓN")
display(importancia_arbol.head(20))

# ============================================================
# 3. MODELOS NO SUPERVISADOS
# ============================================================

variables_cluster_numericas = [
    "SS_INTE",
    "SS_TV",
    "SS_TEL",
    "SS_SVA",
    "SS_MOVIL",
    "SS_DATOS",
    "SS_TV_APP",
    "TOTAL_SERVICIOS",
    "ANTG_MESES_TODO",
    "MONTO_POR_SS",
    "VLOR_NETO_PROM",
    "MESES_COBROS",
    "NUM_RECLAMOS",
    "NUM_RECLAMOS_TECNICO_TERRENO",
    "FLAG_TIENE_RECLAMOS",
    "EVENTOS_DISTINTOS",
    "HORAS_TOTALES_RUT",
    "FLAG_TIENE_EVENTOS_TECNICOS",
    "VELOCIDAD_BAJADA",
    "VELOCIDAD_SUBIDA",
    "LATITUD_DECIMAL",
    "LONGITUD_DECIMAL"
]

variables_cluster_categoricas = [
    "EMPRESA",
    "CRUCE_ACTIVO",
    "CODI_ADMZONAL",
    "REGION",
    "PROVINCIA"
]

variables_cluster_numericas = [
    col for col in variables_cluster_numericas
    if col in df_model.columns
]

variables_cluster_categoricas = [
    col for col in variables_cluster_categoricas
    if col in df_model.columns
]

variables_cluster = variables_cluster_numericas + variables_cluster_categoricas

df_cluster_base = df_model[variables_cluster + [target]].copy()
X_cluster = df_cluster_base.drop(columns=[target])

try:
    encoder_cluster = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
except TypeError:
    encoder_cluster = OneHotEncoder(handle_unknown="ignore", sparse=True)

preprocesamiento_cluster = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), variables_cluster_numericas),
        ("cat", encoder_cluster, variables_cluster_categoricas)
    ],
    remainder="drop"
)

X_cluster_trans = preprocesamiento_cluster.fit_transform(X_cluster)

print("VARIABLES USADAS EN MODELOS NO SUPERVISADOS")
display(pd.DataFrame({
    "tipo": ["numéricas", "categóricas", "total originales"],
    "cantidad": [
        len(variables_cluster_numericas),
        len(variables_cluster_categoricas),
        len(variables_cluster)
    ]
}))

# ------------------------------------------------------------
# 3.1 K-Means
# ------------------------------------------------------------

sample_size = min(8000, X_cluster_trans.shape[0])
rng = np.random.default_rng(42)
sample_idx = rng.choice(X_cluster_trans.shape[0], size=sample_size, replace=False)

X_cluster_sample = X_cluster_trans[sample_idx]

if hasattr(X_cluster_sample, "toarray"):
    X_cluster_sample_dense = X_cluster_sample.toarray()
else:
    X_cluster_sample_dense = X_cluster_sample

resultados_kmeans = []

for k in range(2, 8):
    kmeans_temp = MiniBatchKMeans(
        n_clusters=k,
        random_state=42,
        batch_size=4096,
        n_init=10
    )
    
    labels_temp = kmeans_temp.fit_predict(X_cluster_trans)
    labels_sample = labels_temp[sample_idx]
    
    silhouette = silhouette_score(X_cluster_sample_dense, labels_sample)
    
    resultados_kmeans.append({
        "modelo": "K-Means",
        "k": k,
        "inertia": kmeans_temp.inertia_,
        "silhouette_sample": round(silhouette, 4)
    })

resultados_kmeans_df = pd.DataFrame(resultados_kmeans)

k_kmeans = int(
    resultados_kmeans_df
    .sort_values("silhouette_sample", ascending=False)
    .iloc[0]["k"]
)

modelo_kmeans = MiniBatchKMeans(
    n_clusters=k_kmeans,
    random_state=42,
    batch_size=4096,
    n_init=10
)

labels_kmeans = modelo_kmeans.fit_predict(X_cluster_trans)

df_kmeans = df_cluster_base.copy()
df_kmeans["CLUSTER_KMEANS"] = labels_kmeans

# Distancia al centroide para detección exploratoria de outliers
distancias_kmeans = modelo_kmeans.transform(X_cluster_trans).min(axis=1)
df_kmeans["DISTANCIA_CENTROIDE"] = distancias_kmeans

umbral_outlier = np.percentile(distancias_kmeans, 99)
df_kmeans["OUTLIER_CLUSTER"] = (df_kmeans["DISTANCIA_CENTROIDE"] >= umbral_outlier).astype(int)

perfil_kmeans = (
    df_kmeans
    .groupby("CLUSTER_KMEANS")
    .agg(
        clientes=(target, "count"),
        abandonos=(target, "sum"),
        tasa_abandono=(target, "mean"),
        total_servicios_prom=("TOTAL_SERVICIOS", "mean"),
        antiguedad_prom=("ANTG_MESES_TODO", "mean"),
        monto_por_ss_prom=("MONTO_POR_SS", "mean"),
        valor_neto_prom=("VLOR_NETO_PROM", "mean"),
        reclamos_prom=("NUM_RECLAMOS", "mean"),
        eventos_prom=("EVENTOS_DISTINTOS", "mean"),
        horas_eventos_prom=("HORAS_TOTALES_RUT", "mean"),
        velocidad_bajada_prom=("VELOCIDAD_BAJADA", "mean"),
        velocidad_subida_prom=("VELOCIDAD_SUBIDA", "mean"),
        outliers=("OUTLIER_CLUSTER", "sum")
    )
    .reset_index()
)

perfil_kmeans["porcentaje_clientes"] = (
    perfil_kmeans["clientes"] / len(df_kmeans) * 100
).round(2)

perfil_kmeans["tasa_abandono"] = (
    perfil_kmeans["tasa_abandono"] * 100
).round(2)

perfil_kmeans = perfil_kmeans.sort_values("tasa_abandono", ascending=False)

print("EVALUACIÓN DE K - K-MEANS")
display(resultados_kmeans_df)

print(f"K ÓPTIMO K-MEANS SEGÚN SILHOUETTE: {k_kmeans}")

print("PERFIL GENERAL DE CLUSTERS - K-MEANS")
display(perfil_kmeans)

# ------------------------------------------------------------
# 3.2 Clustering Jerárquico sobre muestra
# ------------------------------------------------------------

df_cluster_sample = df_cluster_base.iloc[sample_idx].copy()

resultados_jerarquico = []

for k in range(2, 8):
    jerarquico_temp = AgglomerativeClustering(
        n_clusters=k,
        linkage="ward"
    )
    
    labels_jerarquico = jerarquico_temp.fit_predict(X_cluster_sample_dense)
    silhouette = silhouette_score(X_cluster_sample_dense, labels_jerarquico)
    
    resultados_jerarquico.append({
        "modelo": "Clustering Jerárquico",
        "k": k,
        "silhouette_sample": round(silhouette, 4)
    })

resultados_jerarquico_df = pd.DataFrame(resultados_jerarquico)

k_jerarquico = int(
    resultados_jerarquico_df
    .sort_values("silhouette_sample", ascending=False)
    .iloc[0]["k"]
)

modelo_jerarquico = AgglomerativeClustering(
    n_clusters=k_jerarquico,
    linkage="ward"
)

labels_jerarquico_final = modelo_jerarquico.fit_predict(X_cluster_sample_dense)

df_jerarquico = df_cluster_sample.copy()
df_jerarquico["CLUSTER_JERARQUICO"] = labels_jerarquico_final

perfil_jerarquico = (
    df_jerarquico
    .groupby("CLUSTER_JERARQUICO")
    .agg(
        clientes=(target, "count"),
        abandonos=(target, "sum"),
        tasa_abandono=(target, "mean"),
        total_servicios_prom=("TOTAL_SERVICIOS", "mean"),
        antiguedad_prom=("ANTG_MESES_TODO", "mean"),
        monto_por_ss_prom=("MONTO_POR_SS", "mean"),
        valor_neto_prom=("VLOR_NETO_PROM", "mean"),
        reclamos_prom=("NUM_RECLAMOS", "mean"),
        eventos_prom=("EVENTOS_DISTINTOS", "mean"),
        horas_eventos_prom=("HORAS_TOTALES_RUT", "mean"),
        velocidad_bajada_prom=("VELOCIDAD_BAJADA", "mean"),
        velocidad_subida_prom=("VELOCIDAD_SUBIDA", "mean")
    )
    .reset_index()
)

perfil_jerarquico["porcentaje_muestra"] = (
    perfil_jerarquico["clientes"] / len(df_jerarquico) * 100
).round(2)

perfil_jerarquico["tasa_abandono"] = (
    perfil_jerarquico["tasa_abandono"] * 100
).round(2)

perfil_jerarquico = perfil_jerarquico.sort_values("tasa_abandono", ascending=False)

print("EVALUACIÓN DE K - CLUSTERING JERÁRQUICO")
display(resultados_jerarquico_df)

print(f"K ÓPTIMO CLUSTERING JERÁRQUICO SEGÚN SILHOUETTE: {k_jerarquico}")

print("PERFIL GENERAL DE CLUSTERS - CLUSTERING JERÁRQUICO")
display(perfil_jerarquico)

# Comparación no supervisada
comparacion_no_supervisada = pd.DataFrame([
    {
        "modelo": "K-Means",
        "base": "dataset completo",
        "k_optimo": k_kmeans,
        "silhouette": resultados_kmeans_df[
            resultados_kmeans_df["k"] == k_kmeans
        ]["silhouette_sample"].iloc[0],
        "observacion": "Segmentación escalable sobre toda la base"
    },
    {
        "modelo": "Clustering Jerárquico",
        "base": f"muestra de {sample_size} registros",
        "k_optimo": k_jerarquico,
        "silhouette": resultados_jerarquico_df[
            resultados_jerarquico_df["k"] == k_jerarquico
        ]["silhouette_sample"].iloc[0],
        "observacion": "Segmentación exploratoria por costo computacional"
    }
])

print("COMPARACIÓN DE MODELOS NO SUPERVISADOS")
display(comparacion_no_supervisada)

# ============================================================
# 4. REGLAS DE ASOCIACIÓN - APRIORI
# ============================================================

# Construcción de matriz transaccional de servicios
df_items = pd.DataFrame(index=df_model.index)

servicios_binarios = {
    "INTERNET": ("SS_INTE", 0),
    "TV": ("SS_TV", 0),
    "TELEFONIA": ("SS_TEL", 0),
    "SVA": ("SS_SVA", 0),
    "TV_APP": ("SS_TV_APP", 0),
    "MOVIL": ("SS_MOVIL", 0),
    "DATOS": ("SS_DATOS", 0),
    "FOX_SPORT": ("FOX_SPORT_VIGENTE", 0),
    "OFFICE": ("OFFICE_VIGENTE", 0),
    "PLAN_HBO": ("PLAN_HBO_VIGENTE", 0),
    "TNT_SPORT": ("TNT_SPORT_VIGENTE", 0),
    "ALARMCOM": ("PROVEEDOR_FINAL_ALARMCOM_VIGENTE", 0)
}

for item, (col, umbral) in servicios_binarios.items():
    if col in df_model.columns:
        df_items[item] = (df_model[col] > umbral).astype(bool)

# Eliminar items sin frecuencia suficiente mínima para evitar reglas irrelevantes
frecuencia_items = df_items.mean().sort_values(ascending=False)
items_validos = frecuencia_items[frecuencia_items >= 0.005].index.tolist()
df_items = df_items[items_validos]

min_support = 0.005
min_confidence = 0.15
max_itemset_size = 3

support_lookup = {}
frequent_itemsets = []

for size in range(1, max_itemset_size + 1):
    for combo in itertools.combinations(df_items.columns, size):
        soporte = df_items[list(combo)].all(axis=1).mean()
        
        if soporte >= min_support:
            combo_key = tuple(sorted(combo))
            support_lookup[combo_key] = soporte
            
            frequent_itemsets.append({
                "itemset": combo_key,
                "tamaño": size,
                "support": soporte
            })

frequent_itemsets_df = pd.DataFrame(frequent_itemsets)

if not frequent_itemsets_df.empty:
    frequent_itemsets_df["support"] = frequent_itemsets_df["support"].round(4)
    frequent_itemsets_df = frequent_itemsets_df.sort_values(
        ["tamaño", "support"],
        ascending=[True, False]
    )

rules = []

for itemset_key, support_itemset in support_lookup.items():
    if len(itemset_key) < 2:
        continue
    
    itemset_set = set(itemset_key)
    
    for r in range(1, len(itemset_key)):
        for antecedente in itertools.combinations(itemset_key, r):
            antecedente_key = tuple(sorted(antecedente))
            consecuente_key = tuple(sorted(itemset_set - set(antecedente)))
            
            support_antecedente = support_lookup.get(antecedente_key)
            support_consecuente = support_lookup.get(consecuente_key)
            
            if support_antecedente is None or support_consecuente is None:
                continue
            
            confianza = support_itemset / support_antecedente
            lift = confianza / support_consecuente
            
            if confianza >= min_confidence:
                mask_antecedente = df_items[list(antecedente_key)].all(axis=1)
                mask_itemset = df_items[list(itemset_key)].all(axis=1)
                
                tasa_abandono_antecedente = df_model.loc[mask_antecedente, target].mean()
                tasa_abandono_itemset = df_model.loc[mask_itemset, target].mean()
                
                rules.append({
                    "antecedente": " + ".join(antecedente_key),
                    "consecuente": " + ".join(consecuente_key),
                    "support": support_itemset,
                    "confidence": confianza,
                    "lift": lift,
                    "tasa_abandono_antecedente": tasa_abandono_antecedente,
                    "tasa_abandono_itemset": tasa_abandono_itemset
                })

rules_df = pd.DataFrame(rules)

if not rules_df.empty:
    rules_df["support"] = rules_df["support"].round(4)
    rules_df["confidence"] = rules_df["confidence"].round(4)
    rules_df["lift"] = rules_df["lift"].round(4)
    rules_df["tasa_abandono_antecedente"] = (
        rules_df["tasa_abandono_antecedente"] * 100
    ).round(2)
    rules_df["tasa_abandono_itemset"] = (
        rules_df["tasa_abandono_itemset"] * 100
    ).round(2)
    
    rules_df = rules_df.sort_values(
        ["lift", "confidence", "support"],
        ascending=False
    )

print("FRECUENCIA DE ITEMS DE SERVICIOS")
display(
    frecuencia_items
    .reset_index()
    .rename(columns={"index": "item", 0: "frecuencia"})
)

print("ITEMSETS FRECUENTES - APRIORI")
display(frequent_itemsets_df.head(30))

print("REGLAS DE ASOCIACIÓN - APRIORI")
display(rules_df.head(30))

# ------------------------------------------------------------
# 5. Guardado de artefactos
# ------------------------------------------------------------

os.makedirs("data/processed", exist_ok=True)

resultados_supervisados_df.to_csv(
    "data/processed/resultados_modelos_supervisados.csv",
    index=False,
    encoding="utf-8-sig"
)

perfil_kmeans.to_csv(
    "data/processed/perfil_clusters_kmeans.csv",
    index=False,
    encoding="utf-8-sig"
)

perfil_jerarquico.to_csv(
    "data/processed/perfil_clusters_jerarquico.csv",
    index=False,
    encoding="utf-8-sig"
)

comparacion_no_supervisada.to_csv(
    "data/processed/comparacion_no_supervisada.csv",
    index=False,
    encoding="utf-8-sig"
)

rules_df.to_csv(
    "data/processed/reglas_asociacion_apriori.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Artefactos de modelamiento guardados en data/processed/")
```

#### Interpretación de la Fase 4: Modelamiento

La fase de modelamiento incorporó tres enfoques complementarios: modelos supervisados para predecir abandono, modelos no supervisados para segmentar clientes y reglas de asociación para descubrir patrones de contratación conjunta de servicios.

En el bloque supervisado se evaluaron cuatro modelos: Regresión Logística balanceada, SVM lineal balanceado, Árbol de Decisión balanceado y Naive Bayes. El conjunto de entrenamiento tuvo **114.115 registros** y el conjunto de prueba **48.907 registros**, manteniendo en ambos casos una tasa de abandono de **3,49%**.

El mejor modelo según `F1-Score` fue la **Regresión Logística balanceada**, con un F1-Score de **0,0962**, Recall de **0,7235**, ROC-AUC de **0,6683** y PR-AUC de **0,0629**. Este modelo logró detectar **1.235 de los 1.707 abandonos reales** del conjunto de prueba. Si bien su precisión es baja, esto es esperable dado el fuerte desbalance de clases. Desde el punto de vista del negocio, el modelo resulta útil porque prioriza clientes en riesgo, aunque genera falsos positivos que deben ser controlados por el costo de campaña.

El SVM lineal obtuvo un desempeño muy similar, con Recall de **0,7299** y F1-Score de **0,0961**. El Árbol de Decisión alcanzó el mayor Recall, con **0,7686**, pero a costa de más falsos positivos y menor F1-Score. Naive Bayes obtuvo una Accuracy alta de **0,9646**, pero prácticamente no detectó abandonos reales, con un Recall de solo **0,0012**. Esto demuestra que Accuracy no es una métrica suficiente en problemas desbalanceados.

La matriz de confusión de la Regresión Logística muestra **472 falsos negativos** y **22.731 falsos positivos**. En el contexto de abandono de clientes, el falso negativo es más crítico, ya que implica no detectar a un cliente que efectivamente abandona. El falso positivo tiene un costo comercial asociado, pero puede ser tolerable si la campaña de retención es de bajo costo o no invasiva. Por ello, el modelo debe utilizarse como herramienta de priorización, no como regla automática de contacto masivo.

```python

# ============================================================
# ANEXOS PARA PRESENTACIÓN
# Código 5: Variables numéricas más asociadas al abandono
# ============================================================

# Carga segura
try:
    df_res
except NameError:
    try:
        df
    except NameError:
        ruta_data = "data/Abandono_Clientes.csv"
        df = pd.read_csv(
            ruta_data,
            sep=";",
            encoding="utf-8-sig",
            low_memory=False
        )
    df_res = df[df["SEGMENTO"] == "RESIDENCIAL"].copy()

target = "ABANDONO_FULL"

# Variables numéricas críticas sin variables de fuga
variables_corr_presentacion = [
    "ANTG_MESES_TODO",
    "VLOR_NETO_PROM",
    "SS_TEL",
    "SS_TV",
    "SS_TV_APP",
    "MESES_COBROS",
    "VELOCIDAD_BAJADA",
    "LATITUD_DECIMAL",
    "SUPERFICIE_KM2",
    "TOTAL_SERVICIOS",
    "MONTO_POR_SS",
    "EVENTOS_DISTINTOS"
]

# TOTAL_SERVICIOS puede no existir en df_res; se crea solo para este gráfico
df_corr_presentacion = df_res.copy()

if "TOTAL_SERVICIOS" not in df_corr_presentacion.columns:
    columnas_servicios = [
        "SS_INTE", "SS_TV", "SS_TEL", "SS_SVA",
        "SS_MOVIL", "SS_DATOS", "SS_TV_APP"
    ]
    columnas_servicios_existentes = [
        col for col in columnas_servicios
        if col in df_corr_presentacion.columns
    ]
    df_corr_presentacion["TOTAL_SERVICIOS"] = (
        df_corr_presentacion[columnas_servicios_existentes].sum(axis=1)
    )

variables_corr_presentacion = [
    col for col in variables_corr_presentacion
    if col in df_corr_presentacion.columns
]

# Imputación temporal solo para correlación exploratoria
df_corr_aux = df_corr_presentacion[[target] + variables_corr_presentacion].copy()

for col in variables_corr_presentacion:
    if df_corr_aux[col].isna().any():
        df_corr_aux[col] = df_corr_aux[col].fillna(df_corr_aux[col].median())

correlaciones = (
    df_corr_aux
    .corr(method="pearson")[target]
    .drop(target)
    .reset_index()
)

correlaciones.columns = ["variable", "correlacion"]
correlaciones["correlacion_abs"] = correlaciones["correlacion"].abs()

correlaciones = correlaciones.sort_values(
    "correlacion_abs",
    ascending=True
).tail(8)

display(correlaciones.sort_values("correlacion_abs", ascending=False))

# Gráfico
plt.figure(figsize=(9, 5.2))

colores = [
    COLOR_ACENTO if valor > 0 else COLOR_SECUNDARIO
    for valor in correlaciones["correlacion"]
]

bars = plt.barh(
    correlaciones["variable"],
    correlaciones["correlacion"],
    color=colores
)

plt.axvline(0, color=COLOR_TEXTO, linewidth=0.8)

plt.title(
    "El abandono no depende de una sola variable",
    loc="left",
    color=COLOR_TEXTO,
    pad=15
)

plt.xlabel("Correlación con ABANDONO_FULL")
plt.ylabel("")

# Etiquetas directas
for bar, valor in zip(bars, correlaciones["correlacion"]):
    desplazamiento = 0.003 if valor >= 0 else -0.003
    alineacion = "left" if valor >= 0 else "right"
    
    plt.text(
        valor + desplazamiento,
        bar.get_y() + bar.get_height() / 2,
        f"{valor:.3f}".replace(".", ","),
        va="center",
        ha=alineacion,
        fontsize=10,
        color=COLOR_TEXTO
    )

plt.xlim(-0.08, 0.06)

guardar_grafico("04_correlaciones_abandono.png")
```

```python

# ============================================================
# ANEXOS PARA PRESENTACIÓN
# Código 9: Reglas de asociación Apriori para storytelling
# ============================================================

# Carga segura de reglas Apriori
try:
    rules_df
except NameError:
    ruta_reglas = "data/processed/reglas_asociacion_apriori.csv"
    rules_df = pd.read_csv(ruta_reglas, encoding="utf-8-sig")

reglas = rules_df.copy()

# Filtrar reglas menos útiles para presentación:
# INTERNET aparece en todos los clientes, por lo que no aporta discriminación comercial.
reglas_presentacion = reglas[
    ~reglas["antecedente"].str.contains("INTERNET", case=False, na=False) &
    ~reglas["consecuente"].str.contains("INTERNET", case=False, na=False)
].copy()

# Evitar reglas demasiado extensas para lámina
reglas_presentacion["largo_regla"] = (
    reglas_presentacion["antecedente"].str.count(r"\+") +
    reglas_presentacion["consecuente"].str.count(r"\+") +
    2
)

reglas_presentacion = reglas_presentacion[
    reglas_presentacion["largo_regla"] <= 3
].copy()

# Seleccionar reglas con buen lift, confianza y soporte mínimo
reglas_presentacion = reglas_presentacion[
    (reglas_presentacion["support"] >= 0.005) &
    (reglas_presentacion["confidence"] >= 0.15) &
    (reglas_presentacion["lift"] > 1)
].copy()

# Ordenar por lift, pero mantener solo las más claras para storytelling
reglas_presentacion = (
    reglas_presentacion
    .sort_values(["lift", "confidence", "support"], ascending=False)
    .head(7)
    .copy()
)

reglas_presentacion["regla"] = (
    reglas_presentacion["antecedente"] +
    " → " +
    reglas_presentacion["consecuente"]
)

# Tabla ejecutiva
tabla_reglas_presentacion = reglas_presentacion[[
    "regla",
    "support",
    "confidence",
    "lift",
    "tasa_abandono_itemset"
]].copy()

tabla_reglas_presentacion["support"] = (
    tabla_reglas_presentacion["support"] * 100
).round(2)

tabla_reglas_presentacion["confidence"] = (
    tabla_reglas_presentacion["confidence"] * 100
).round(2)

tabla_reglas_presentacion["lift"] = (
    tabla_reglas_presentacion["lift"]
).round(2)

tabla_reglas_presentacion = tabla_reglas_presentacion.rename(columns={
    "regla": "Regla",
    "support": "Soporte (%)",
    "confidence": "Confianza (%)",
    "lift": "Lift",
    "tasa_abandono_itemset": "Abandono grupo (%)"
})

display(tabla_reglas_presentacion)

# Gráfico: Lift de reglas principales
reglas_plot = reglas_presentacion.sort_values("lift", ascending=True)

plt.figure(figsize=(10, 5.5))

bars = plt.barh(
    reglas_plot["regla"],
    reglas_plot["lift"],
    color=[
        COLOR_ACENTO if i == len(reglas_plot) - 1 else COLOR_BASE
        for i in range(len(reglas_plot))
    ],
    height=0.55
)

plt.title(
    "Algunas combinaciones de servicios funcionan como anclas de fidelización",
    loc="left",
    color=COLOR_TEXTO,
    pad=15
)

plt.xlabel("Lift de la regla")
plt.ylabel("")
plt.xlim(0, reglas_plot["lift"].max() * 1.25)

# Línea de referencia: lift = 1
plt.axvline(
    1,
    color=COLOR_TEXTO,
    linewidth=0.8,
    linestyle="--"
)

plt.text(
    1.03,
    -0.75,
    "Lift = 1: sin asociación",
    fontsize=9,
    color=COLOR_TEXTO
)

# Etiquetas directas con confianza
for bar, lift, confianza in zip(
    bars,
    reglas_plot["lift"],
    reglas_plot["confidence"]
):
    plt.text(
        bar.get_width() + 0.08,
        bar.get_y() + bar.get_height() / 2,
        f"Lift {lift:.2f} | Conf. {confianza*100:.1f}%".replace(".", ","),
        va="center",
        fontsize=10,
        color=COLOR_TEXTO
    )

plt.gca().spines["left"].set_visible(False)
plt.tick_params(axis="y", length=0)
plt.grid(axis="x", alpha=0.15)

guardar_grafico("08_reglas_apriori_lift.png")
```

```python

# ============================================================
# ANEXOS PARA PRESENTACIÓN
# Código 10: Campañas estratégicas combinando modelos
# ============================================================

# Requiere haber ejecutado previamente la Fase 4 completa:
# - df_model
# - modelos_entrenados
# - df_kmeans o labels_kmeans

target = "ABANDONO_FULL"

# ------------------------------------------------------------
# 1. Base para plan de acción
# ------------------------------------------------------------

try:
    df_model
except NameError:
    ruta_preparado = "data/processed/clientes_residenciales_preparado.csv"
    df_model = pd.read_csv(ruta_preparado, encoding="utf-8-sig")

df_accion = df_model.copy()

# ------------------------------------------------------------
# 2. Incorporar cluster K-Means
# ------------------------------------------------------------

if "CLUSTER_KMEANS" in df_accion.columns:
    pass
elif "df_kmeans" in globals() and len(df_kmeans) == len(df_accion):
    df_accion["CLUSTER_KMEANS"] = df_kmeans["CLUSTER_KMEANS"].values
elif "labels_kmeans" in globals() and len(labels_kmeans) == len(df_accion):
    df_accion["CLUSTER_KMEANS"] = labels_kmeans
else:
    raise NameError(
        "No se encontraron clusters K-Means. Ejecuta primero la sección de K-Means de la Fase 4."
    )

nombres_clusters = {
    0: "BA menos consolidados",
    1: "Consolidados y paquetizados",
    2: "Territorial diferenciado"
}

df_accion["NOMBRE_CLUSTER"] = (
    df_accion["CLUSTER_KMEANS"]
    .map(nombres_clusters)
    .fillna("Cluster " + df_accion["CLUSTER_KMEANS"].astype(str))
)

# ------------------------------------------------------------
# 3. Incorporar score de riesgo del modelo supervisado ganador
# ------------------------------------------------------------

if "modelos_entrenados" not in globals():
    raise NameError(
        "No se encontraron modelos entrenados. Ejecuta primero la sección supervisada de la Fase 4."
    )

nombre_modelo_accion = "Regresión logística balanceada"

pipeline_accion = modelos_entrenados[nombre_modelo_accion]["pipeline"]

X_accion = df_accion.drop(columns=[target], errors="ignore")

# Evitar que columnas agregadas para análisis entren a la predicción
columnas_no_modelo = [
    "CLUSTER_KMEANS",
    "NOMBRE_CLUSTER",
    "SCORE_RIESGO",
    "BANDA_RIESGO"
]

X_accion = X_accion.drop(
    columns=[col for col in columnas_no_modelo if col in X_accion.columns],
    errors="ignore"
)

df_accion["SCORE_RIESGO"] = pipeline_accion.predict_proba(X_accion)[:, 1]

umbral_top_20 = df_accion["SCORE_RIESGO"].quantile(0.80)
umbral_top_10 = df_accion["SCORE_RIESGO"].quantile(0.90)

df_accion["BANDA_RIESGO"] = np.select(
    [
        df_accion["SCORE_RIESGO"] >= umbral_top_10,
        df_accion["SCORE_RIESGO"] >= umbral_top_20
    ],
    [
        "Muy alto riesgo",
        "Alto riesgo"
    ],
    default="Riesgo medio/bajo"
)

# ------------------------------------------------------------
# 4. Variables de servicios para campañas
# ------------------------------------------------------------

def flag_servicio(columna):
    if columna in df_accion.columns:
        return df_accion[columna].fillna(0) > 0
    return pd.Series(False, index=df_accion.index)

df_accion["TIENE_TV"] = flag_servicio("SS_TV")
df_accion["TIENE_TELEFONIA"] = flag_servicio("SS_TEL")
df_accion["TIENE_SVA"] = flag_servicio("SS_SVA")
df_accion["TIENE_TNT"] = flag_servicio("TNT_SPORT_VIGENTE")
df_accion["TIENE_ALARMCOM"] = flag_servicio("PROVEEDOR_FINAL_ALARMCOM_VIGENTE")

alto_riesgo = df_accion["SCORE_RIESGO"] >= umbral_top_20
muy_alto_riesgo = df_accion["SCORE_RIESGO"] >= umbral_top_10

# ------------------------------------------------------------
# 5. Definición de campañas accionables
# ------------------------------------------------------------

campanias = {
    "Pack base para BA en riesgo": {
        "descripcion": "Clientes del cluster BA menos consolidado, alto riesgo y sin TV.",
        "mascara": (
            (df_accion["CLUSTER_KMEANS"] == 0) &
            alto_riesgo &
            (~df_accion["TIENE_TV"])
        ),
        "accion": "Ofrecer pack Internet + TV o upgrade de paquetización.",
        "modelo_usado": "Regresión Logística + K-Means",
        "regla_apriori": "TELEFONIA → TV / TNT_SPORT → TV"
    },
    "Cross-selling Telefonía a TV": {
        "descripcion": "Clientes con Telefonía, alto riesgo y sin TV.",
        "mascara": (
            df_accion["TIENE_TELEFONIA"] &
            alto_riesgo &
            (~df_accion["TIENE_TV"])
        ),
        "accion": "Ofrecer TV como segundo servicio ancla.",
        "modelo_usado": "Regresión Logística + Apriori",
        "regla_apriori": "TELEFONIA → TV"
    },
    "Ancla deportiva para clientes TV": {
        "descripcion": "Clientes con TV, alto riesgo y sin TNT Sport.",
        "mascara": (
            df_accion["TIENE_TV"] &
            alto_riesgo &
            (~df_accion["TIENE_TNT"])
        ),
        "accion": "Ofrecer TNT Sport como beneficio de permanencia.",
        "modelo_usado": "Regresión Logística + Apriori",
        "regla_apriori": "TNT_SPORT → TV"
    },
    "Protección de clientes consolidados": {
        "descripcion": "Clientes paquetizados, alto valor y riesgo muy alto.",
        "mascara": (
            (df_accion["CLUSTER_KMEANS"] == 1) &
            muy_alto_riesgo
        ),
        "accion": "Aplicar beneficio preventivo de permanencia, no descuento masivo.",
        "modelo_usado": "Regresión Logística + K-Means",
        "regla_apriori": "No aplica directamente"
    }
}

# ------------------------------------------------------------
# 6. Resumen ejecutivo de campañas
# ------------------------------------------------------------

resumen_campanias = []

for nombre, info in campanias.items():
    mask = info["mascara"]
    base = df_accion.loc[mask].copy()
    
    if len(base) == 0:
        continue
    
    resumen_campanias.append({
        "Campaña": nombre,
        "Clientes priorizables": len(base),
        "% base residencial": len(base) / len(df_accion) * 100,
        "Abandono real histórico (%)": base[target].mean() * 100,
        "Score riesgo promedio (%)": base["SCORE_RIESGO"].mean() * 100,
        "Servicios promedio": base["TOTAL_SERVICIOS"].mean() if "TOTAL_SERVICIOS" in base.columns else np.nan,
        "Cluster dominante": base["NOMBRE_CLUSTER"].mode().iloc[0],
        "Modelo usado": info["modelo_usado"],
        "Regla Apriori": info["regla_apriori"],
        "Acción recomendada": info["accion"]
    })

tabla_campanias = pd.DataFrame(resumen_campanias)

tabla_campanias["% base residencial"] = tabla_campanias["% base residencial"].round(2)
tabla_campanias["Abandono real histórico (%)"] = tabla_campanias["Abandono real histórico (%)"].round(2)
tabla_campanias["Score riesgo promedio (%)"] = tabla_campanias["Score riesgo promedio (%)"].round(2)
tabla_campanias["Servicios promedio"] = tabla_campanias["Servicios promedio"].round(2)

tabla_campanias = tabla_campanias.sort_values(
    "Clientes priorizables",
    ascending=False
)

display(tabla_campanias)

# ------------------------------------------------------------
# 7. Gráfico ejecutivo: tamaño de oportunidad por campaña
# ------------------------------------------------------------

tabla_plot = tabla_campanias.sort_values("Clientes priorizables", ascending=True)

plt.figure(figsize=(10, 5.5))

bars = plt.barh(
    tabla_plot["Campaña"],
    tabla_plot["Clientes priorizables"],
    color=[
        COLOR_ACENTO if valor == tabla_plot["Clientes priorizables"].max() else COLOR_BASE
        for valor in tabla_plot["Clientes priorizables"]
    ],
    height=0.55
)

plt.title(
    "El modelo permite transformar riesgo en campañas concretas",
    loc="left",
    color=COLOR_TEXTO,
    pad=15
)

plt.xlabel("Clientes priorizables")
plt.ylabel("")

for bar, clientes, riesgo in zip(
    bars,
    tabla_plot["Clientes priorizables"],
    tabla_plot["Score riesgo promedio (%)"]
):
    plt.text(
        bar.get_width() + tabla_plot["Clientes priorizables"].max() * 0.015,
        bar.get_y() + bar.get_height() / 2,
        f"{formato_miles(clientes)} clientes | riesgo prom. {riesgo:.1f}%".replace(".", ","),
        va="center",
        fontsize=10,
        color=COLOR_TEXTO
    )

plt.xlim(0, tabla_plot["Clientes priorizables"].max() * 1.30)
plt.gca().spines["left"].set_visible(False)
plt.tick_params(axis="y", length=0)
plt.grid(axis="x", alpha=0.15)

guardar_grafico("09_campanias_estrategicas.png")

# Guardar tabla para usarla como apoyo en la presentación
tabla_campanias.to_csv(
    "outputs/presentacion/tabla_campanias_estrategicas.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Tabla de campañas guardada en outputs/presentacion/tabla_campanias_estrategicas.csv")
```
